```python
import os
import numpy as np
import torch
import json
import re
import matplotlib.pyplot as plt
```

```python
def get_accuracies_from_raw_logs(log_path):
    top1s = []; top5s = []
    with open(log_path, 'r') as handle:
        for line in handle:
            if '* Acc@1' not in line: continue
            acc1, acc5 = re.findall("\d+\.\d+", line)
            top1s.append(float(acc1))
            top5s.append(float(acc5))
    return top1s, top5s
                
```

```python
reversed_attn_path = '/srv/share4/gstoica3/checkpoints/swin_reversedBase_transformer_cifar/swin_tiny_patch4_window7_224_cifar100/correct/log_rank0.txt'
regular_attn_path = '/srv/share4/gstoica3/checkpoints/swin_transformer_cifar/swin_tiny_patch4_window7_224_cifar100/baseline_fp32/log_rank0.txt'
```

```python
reversed_top1, reversed_top5 = get_accuracies_from_raw_logs(reversed_attn_path)
reg_top1, reg_top5 = get_accuracies_from_raw_logs(regular_attn_path)
```

```python
plt.plot(list(range(301)), reversed_top1[:301], 'b', label='Reversed Attention')
plt.plot(list(range(301)), reg_top1, 'r', label='Regular Attention')
plt.title('Regular vs Reversed Attention Top 1 Accuracies over Val Set')
plt.show()
```

```python
plt.plot(list(range(301)), reversed_top5[:301], 'b', label='Reversed Attention')
plt.plot(list(range(301)), reg_top5, 'r', label='Regular Attention')
plt.title('Regular vs Reversed Attention Top 5 Accuracies over Val Set')
plt.show()
```
